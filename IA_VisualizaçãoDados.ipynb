**Hipótese**

Quanto maior a presença nas aulas, maior será a nota do aluno.

```python
df_mental_health = df['exam_score']
df_mental_health.describe()
```

```python
df_exam_mental_health = df[(df['attendance_percentage'] < 60) & (df['exam_score'] < 5)]
df.plot.scatter(
    x='exam_score',
    y='attendance_percentage',
    color='blue',
    alpha=1,
    title='Relação entre Presença e Notas',
    xlabel='Nota',
    ylabel='Presença'
)
```

**Observações:**

- A maioria dos pontos se concentra em presenças acima de 80%, com uma grande variação de notas, inclusive notas baixas.

- Estudantes com presenças muito baixas (abaixo de 70%) tendem a ter notas mais baixas, mas o oposto não é garantido — alta presença não implica necessariamente em alto desempenho.

- A dispersão é grande, e não se observa uma relação clara ou forte de proporcionalidade direta.

**Conclusão:**

Apesar da expectativa de que frequência alta traga melhor desempenho, o gráfico não mostra uma correlação forte entre presença e nota. Isso indica que outros fatores — como métodos de estudo, saúde mental ou hábitos fora da sala — podem ter impacto maior nas notas dos alunos.
