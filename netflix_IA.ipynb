## 5. Coluna 'date_added' para datetime

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

## 6. Cria as colunas 'month_added' e 'year_added' a partir de 'date_added'

```python
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year
```

## 7. Títulos por tipo (Movie vs TV Show):

```python
tipo_counts = df['type'].value_counts()
tipo_counts
```

## 8. Filtro por filmes

```python
df_filmes = df[df['type'] == 'Movie'].copy()
df_filmes['duration_int'] = df_filmes['duration'].str.extract('(\d+)').astype(float)
df_filmes_filtrados = df_filmes[
    (df_filmes['duration_int'] > 100) &
    (df_filmes['release_year'] >= 2020)
]
df_filmes
```

## 9. Separa a coluna 'duration' em 'duration_int' (numérico) e 'duration_unit'
(unidade)

```python
df['duration_int'] = df['duration'].str.extract('(\d+)').astype(float)
df['duration_unit'] = df['duration'].str.extract('(\D+)$')
```

## 10. Cria a coluna ‘seasons_int’

```python
df['seasons_int'] = df.apply(
    lambda x: x['duration_int'] if x['type'] == 'TV Show' else None,
    axis=1
)
```

## 11. Série com mais temporadas e o filme com maior tempo de duração

```python
serie_mais_temporadas = df[df['type'] == 'TV Show'].sort_values(by='seasons_int', ascending=False).head(1)
filme_maior_duracao = df[df['type'] == 'Movie'].sort_values(by='duration_int', ascending=False).head(1)
print(serie_mais_temporadas)
print(filme_maior_duracao)
```
